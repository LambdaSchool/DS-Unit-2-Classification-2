```python
picks = browse.sample(40)
```

```python
#Simulate wieghted coin flips whether the loan will default,
#based on Lending Clubs expected default rate
picks['default'] = picks['exp_default_rate'] > np.random.rand(len(picks)) *100
picks['default'].value_counts()
```

```python
picks['paid'] = picks['best']
picks.loc[picks['default'], 'paid'] = picks['worst']
```

```python
columns += ['default', 'paid']
```

```python
picks[columns]
```

```python
picks['paid'].sum() # gives the net profit of the loans based
#on the simulation accounting for the defaults
```

```python
from tqdm import tnrange, tqdm_notebook
from time import sleep

simulations = []
for i in tnrange(1000):
    sleep(0.01)
    picks['default'] = picks['exp_default_rate'] > np.random.rand(len(picks)) *100
    picks['paid'] = picks['best']
    picks.loc[picks['default'], 'paid'] = picks['worst']
    paid = picks['paid'].sum()
    simulations.append(paid)
```

```python
#Showing multiple "Monte Carlo" like simulation results into 
#a pandas series
```

```python
pd.Series(simulations).describe()
```

```python
%matplotlib inline
import seaborn as sns
sns.distplot(simulations)
```

```python
def a_to_d_done(df):
    a_to_d = df['grade'].isin(['A', 'B', 'C', 'D'])
    done = df['loan_status'].isin(['Fully Paid', 'Charged Off'])
    return a_to_d & done

df = loans2019q1.where(a_to_d_done).dropna(subset=['grade', 'loan_status'])
```

```python
df.head()
```

```python
files = glob('LoanStats*.csv')
```

```python
files
```

```python


files = glob('LoanStats*.csv')

dfs = (pd.read_csv(file, skiprows=1, skipfooter=2, engine='python')
       .where(a_to_d_done)
       .dropna(subset=['grade', 'loan_status'])
       for file in tqdm_notebook(files))

df = pd.concat(dfs)
df.shape

```

```python
df.info()
```

```python
#will pull a fraction of the data
#similar to train,test,split
df.sample(frac=0.10)
```

```python
from sklearn.model_selection import train_test_split
#stratify method will keep the same fraction of labels in the split
#so the label you are predicting will have a very close proportion
# of the same label split
#good for class imbalances
subset, _ = train_test_split(df.copy(), train_size=0.10,
                            stratify=df['loan_status'])
```
