```python
browse['worst'] = -25
browse['best'] = (25  # original investment returned, plus interest...
                  + 25 * browse['term']/12 # for a given number of years ...
                  * (browse['int_rate'] - browse['service_fee_rate']) / 100)
```

```python
columns = ['term', 'int_rate', 'exp_default_rate', 'service_fee_rate', 
           'best', 'worst']
browse[columns].sort_values(by='best', ascending=False)
#if I collect for 5 years, I double my money
#lowest upside is a 3 year loan with a 6.46 int_rate
```

```python
#how to use expected default rate, to simulate - pick random number from 1-100, if 1 or 2, say defaulted, greater than 2
#say paid off, what would be my expected return
picks = browse.sample(40)
```

```python
import numpy as np
#Simulate weighted coin flips, whether the loan will default
#based on Lending Club's expected default rate
picks['default'] = picks['exp_default_rate'] > np.random.rand(len(picks))
picks['default'].value_counts()
```

```python
picks['paid'] = picks['best']
picks.loc[picks['default'],'paid'] = picks['worst']
```

```python
columns += ['default', 'paid']
```

```python
picks['paid'].sum()
```

```python
from tqdm import tnrange, tqdm_notebook #progress bar library
from time import sleep

simulations = []
for _ in range(10000): 
    picks['default'] = picks['exp_default_rate'] > np.random.rand(len(picks)) * 100
    picks['paid'] = picks['best']
    picks.loc[picks['default'], 'paid'] = picks['worst']
    paid = picks['paid'].sum()
    simulations.append(paid)
```

```python
pd.Series(simulations).describe()
#in this simulation, we see that it is unlikely we would loose 1000 but we could loose $200, or up $400
```

```python
%matplotlib inline
import seaborn as sns
sns.distplot(simulations);
```

```python
def a_to_d_done(df):
    a_to_d = df['grade'].isin(['A', 'B', 'C', 'D'])
    done = df['loan_status'].isin(['Fully Paid', 'Charged Off'])
    return a_to_d & done

df = loans2019q1.where(a_to_d_done).dropna(subset=['grade', 'loan_status'])
#verify if I do a value_count of grade
df['grade'].value_counts(dropna=False)
```

```python
from glob import glob #get the glob function inside module glob, otherwise you import glob, you have to glob.glob
glob('LoanStats*.csv')
```

```python
files = glob('LoanStats*.csv')

dfs = (pd.read_csv(file, skiprows=1, skipfooter=2, engine='python')
       .where(a_to_d_done)
       .dropna(subset=['grade', 'loan_status'])
       for file in tqdm_notebook(files))

df = pd.concat(dfs)
df.shape
```
