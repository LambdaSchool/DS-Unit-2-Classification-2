```python
# What if found 40 loans like id 152966674, and invest $25 in each ?
# We'd expect ~1 to default and 39 to be fully paid
1 * -25 + 39 * 0.0756 * 25
```

```python
browse.sort_values(by='int_rate', ascending=False).head()
```

```python
browse['worst'] = -25
browse['best'] = (25 #original investment returned, plus interest...
                  + 25 * browse['term']/12 #for a given number of years since interest is paid anually
                  * (browse['int_rate'] - browse['service_fee_rate'])/100)
```

```python
columns = ['term', 'int_rate', 'exp_default_rate', 'service_fee_rate',
          'best', 'worst']
browse[columns].sort_values(by='best', ascending=False)
```

```python
picks = browse.sample(40)
```

```python
import numpy as np
# simulate weighted coin flips, whether the loan will default,
# based on Lending Club's expected default rate
dictionary2.loc['expDefaultRate']
```

```python
np.random.rand(len(picks))*100
```

```python
picks['default'] = picks['exp_default_rate'] > np.random.rand(len(picks)) * 100
picks['default'].value_counts()
```

```python
picks['default']
```

```python
picks.loc[picks['default'], 'paid']
```

```python
picks['paid'] = picks['best']
picks.loc[picks['default'], 'paid'] = picks['worst']
```

```python
columns += ['default', 'paid']
```

```python
picks['paid'].sum()
```

```python
from tqdm import tnrange, tqdm_notebook
from time import sleep

simulations = []
for _ in tnrange(10000):
    picks['default'] = picks['exp_default_rate'] > np.random.rand(len(picks)) * 100
    picks['paid'] = picks['best']
    picks.loc[picks['default'], 'paid'] = picks['worst']
    paid = picks['paid'].sum()
    simulations.append(paid)
```

```python
pd.Series(simulations).describe()
```

```python
%matplotlib inline
import seaborn as sns
sns.distplot(simulations);
```

```python
def a_to_d_done(df):
    a_to_d = df['grade'].isin(['A', 'B', 'C', 'D'])
    done = df['loan_status'].isin(['Fully Paid', 'Charged Off'])
    return a_to_d & done
```

```python
df = loans2019q1.where(a_to_d_done).dropna(subset=['grade', 'loan_status'])
```

```python
df.shape
```

```python
df.loan_status.value_counts(dropna=False)
```

```python
df.grade.value_counts(dropna=False)
```

```python
from glob import glob
```

```python
files = glob('LoanStats*.csv')

dfs = (pd.read_csv(file, skiprows=1, skipfooter=2, engine='python')
       .where(a_to_d_done)
       .dropna(subset=['grade', 'loan_status'])
       for file in tqdm_notebook(files))

df = pd.concat(dfs)
df.shape
```

```python
df.info()
```

```python
df['loan_status'].value_counts(normalize=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
subset, _ = train_test_split(df.copy(), train_size=.10,
                            stratify=df['loan_status'])
```
